```python
from tdc.multi_pred import DTI
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
name = "DAVIS"
data_dti = DTI(name = name,path = "../data")
df = data_dti.get_data()
df['dlen'] = df['Drug'].str.len()
df['tlen'] = df['Target'].str.len()
df.to_csv(f"../data/{name}_seq_len.csv")
drug = df['dlen'].values
target = df['tlen'].values

f,axes = plt.subplots(1,2,figsize=(14,5))

sub_graph = axes[0]
sub_graph.hist(drug, bins= 15,color='skyblue', alpha=0.8)
sub_graph.set_title(f"drug sequence on {name} dataset")
sub_graph.set_xlabel("length")
sub_graph.set_ylabel("count")

sub_graph = axes[1]
sub_graph.hist(target, bins=15,color='skyblue', alpha=0.8)
sub_graph.set_title(f"target sequence on {name} dataset")
sub_graph.set_xlabel("length")
sub_graph.set_ylabel("count")
```
