```python
from transformers import BertTokenizer, BertModel
```

```python

tokenizer = BertTokenizer.from_pretrained('../data/BERT')
model = BertModel.from_pretrained("../data/BERT")
drug = "COc1cc(Nc2ncc(F)c(Nc3ccc4c(n3)NC(=O)C(C)(C)O4)n2)cc(OC)c1OC.O=S(=O)(O)c1ccccc1"
print(f"drug len: {len(drug)}")
encoded_input = tokenizer(drug, return_tensors='pt')
output = model(**encoded_input)
print(output[0].shape)
print(output[1].shape)  ## representation of the whole sentence
print(output[1]) ## representation of the whole sentence
```
