```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/times.ttf')
plt.rc('font',family='Times New Roman')
```

```python
EF5=[2.0,2.0,5.2,6.4,6.4,6.8,7.2,7.6,8.0]
EF10=[1.0,1.0,2.2,3.2,3.2,3.2,3.2,3.4,3.8]
Hit5=[1,1,1,1,1,2,2,3,4]
Hit10=[2,2,2,2,3,5,6,6,8]

x=["KronRLS","SimBoost", "DeepDTA", "GraphDTA", "MccDTI", "MvGraphDTA", "SSGraphDTI", "DTIAM", "HMBVIP"] 
plt.rcParams.update({'font.size': 14})
plt.subplots(2,2,figsize=(10,10),dpi=300)

plt.subplot(2,2,1)
plt.barh(x, EF5, color='royalblue',alpha=0.7)
# plt.title("Attention Weights Across Views")
plt.xlabel("EF5%")
# plt.xlabel("Method")

plt.subplot(2,2,2)
plt.barh(x, EF10, color='royalblue',alpha=0.7)
# plt.title("Attention Weights Across Views")
# plt.xlabel("Method")
plt.xlabel("EF10%")

plt.subplot(2,2,3)
plt.barh(x, Hit5,color='skyblue',alpha=0.7)
# plt.title("Attention Weights Across Views")
# plt.xlabel("Method")
plt.xlabel("Hit@5")

plt.subplot(2,2,4)
plt.barh(x, Hit10, color='skyblue',alpha=0.7)
# plt.title("Attention Weights Across Views")
# plt.xlabel("Method")
plt.xlabel("Hit@10")

plt.tight_layout()
plt.savefig("EF.svg")
plt.savefig("EF.jpg")
plt.show()
```
