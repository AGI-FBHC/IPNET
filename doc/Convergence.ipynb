```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
davis = pd.read_csv("./DAVIS_loss.csv")['loss'].values
kiba = pd.read_csv("./KIBA_loss.csv")['loss'].values
def sampling(data, step=20):
    output=[]
    for i in range(len(data)):
        if i % step == 0:
            output.append(data[i])
    return output

davis = sampling(davis, step=20)
kiba = sampling(kiba, step=20)
```

```python
f, axes = plt.subplots(1, 2, figsize=(10,5))

sub_graph = axes[0]
data = davis
sub_graph.plot(data, label='Loss',color='deepskyblue')
sub_graph.set_title("Training on DAVIS Dataset")
sub_graph.set_xlabel("Epoch")
sub_graph.set_ylabel("Loss")
sub_graph.legend()

sub_graph = axes[1]
data = kiba
sub_graph.plot(data, label='Loss',color='yellowgreen')
sub_graph.set_title("Training on KIAB Dataset")
sub_graph.set_xlabel("Epoch")
sub_graph.set_ylabel("Loss")
sub_graph.legend()

plt.savefig("Convergence.svg")
plt.savefig("Convergence.jpg")
```
